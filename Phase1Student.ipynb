## 📊 Feature 1: Overall Academic Result

 Description:

The overall_result feature classifies students based on their combined performance in all three core subjects: Maths, Science, and Social.
	•	A student is marked “Pass” only if they score 35% or above in all three subjects.
	•	If they score below 35% in any one subject, they are marked as “Fail”.

```python
df1['overall_result'] = (
    (df1['Maths_Percentage'] >= 35) & 
    (df1['Science_Percentage'] >= 35) & 
    (df1['Social_Percentage'] >= 35)
).map({True: 'Pass', False: 'Fail'})

df1['overall_result'].value_counts().plot(kind='bar', color='green')
plt.title("Overall Pass/Fail Count")
plt.xlabel("Result")
plt.ylabel("Count")
plt.show()
```

## 📊 Feature 2: Count of Strong Subjects per Student

Description:

This feature represents the number of core subjects in which a student has scored 75% or more, indicating strong academic performance. The subjects considered are Maths, Science, and Social Studies.

	•	A score ≥ 75% in a subject is considered strong.
	•	The value of strong_subjects_count ranges from 0 to 3:
	•	0: No strong subjects
	•	1: Strong in one subject
	•	2: Strong in two subjects
	•	3: Strong in all three subjects


```python
df1['strong_subjects_count'] = (
    (df1['Maths_Percentage'] >= 75).astype(int) +
    (df1['Science_Percentage'] >= 75).astype(int) +
    (df1['Social_Percentage'] >= 75).astype(int)
)

df1['strong_subjects_count'].value_counts().sort_index().plot(kind='bar', color='orange')
plt.title("Number of Strong Subjects per Student")
plt.xlabel("Number of Strong Subjects")
plt.ylabel("Number of Students")
plt.show()
```

## 📊Feature 3 : Study Support Score (Tuition & Internet)

This feature, study_support_score, is an engineered metric that quantifies the level of academic support a student receives outside of regular school hours. It combines the usage of:
	
	•	Private Tuition – Indicates if the student is attending any paid or structured extra coaching.
	•	Internet Usage – Indicates if the student uses the internet for educational purposes.


The feature is computed as:
	
	•	0 → No tuition and no internet support
	•	1 → Either tuition or internet is used
	•	2 → Both tuition and internet are used


```python
df1['study_support_score'] = (
    (df1['private_tuition'] != 'No').astype(int) +
    (df1['use_internet'] != 'No').astype(int)
)
```

```python
df1['total_score'] = df1['Maths_Percentage'] + df1['Science_Percentage'] + df1['Social_Percentage']
df1['average_score'] = df1['total_score'] / 3
```

```python
avg_scores = df1.groupby('study_support_score')['average_score'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(x='study_support_score', y='average_score', data=avg_scores, palette='mako',hue='average_score')
plt.title("Average Marks vs Study Support (Tuition & Internet)")
plt.xlabel("Study Support Score (0=None, 1=Either, 2=Both)")
plt.ylabel("Average Marks")
plt.tight_layout()
plt.show()
```
