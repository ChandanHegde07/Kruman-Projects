```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("datasets/nas-pupil-marks.csv")
```

```python
df1.head()
```

```python
print("Rows:", df1.shape[0], "Columns:", df1.shape[1])
df1.info()

df1.isnull().sum()
```

```python
df1 = df1.drop_duplicates()
df1.head()
```

```python
df1 = df1.dropna()
df1.head(10)
```

```python
df1.columns = df1.columns.str.strip().str.replace(" ", "_").str.replace("%", "Percent")
df1.head()
```

```python
df1 = df1.dropna(thresh=5)
df1.head()
```

```python
print(df1['STUID'].nunique(), len(df1))
```

```python
df1.columns = df1.columns.str.strip().str.lower().str.replace(' ', '_')
df1.head()
```

```python
df1['age'].value_counts()
```

```python
df1['gender'].value_counts()
```

```python
df1.rename(columns={"maths_percent": "Maths_Percentage"}, inplace=True)
df1.rename(columns={"reading_percent": "Reading_Percentage"}, inplace=True)
df1.rename(columns={"science_percent": "Science_Percentage"}, inplace=True)
df1.rename(columns={"social_percent": "Social_Percentage"}, inplace=True)
df1.head()
```

```python
print(df1.columns.tolist())
```

```python
df1['use_computer'] = df1['use_computer'].fillna('Unknown')
df1['use_calculator'] = df1['use_calculator'].fillna('Unknown')
df1['use_internet'] = df1['use_internet'].fillna('Unknown')
df1['use_dictionary'] = df1['use_dictionary'].fillna('Unknown')
```

```python
df1.tail(1)
```

```python
df1.isnull().values.any()
df1.isnull().sum()
df1.isnull().sum().sum()
```

```python
df1.info()
df1.describe(include='all')
```

```python
df1.dropna(inplace=True)
```

```python
df1.isnull().sum().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df1["Maths_Percentage"], bins=20, color='orange', kde=True)
plt.title("Distribution of Maths Percent")
plt.xlabel("Maths Percent")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df1["Reading_Percentage"], bins=20,color='green', kde=True)
plt.title("Distribution of Reading Percent")
plt.xlabel("Reading Percent")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df1["Science_Percentage"], bins=20,color='red', kde=True)
plt.title("Distribution of Science Percent")
plt.xlabel("Science Percent")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df1["Social_Percentage"], bins=20,color='gray', kde=True)
plt.title("Distribution of Social Percent")
plt.xlabel("Social Percent")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='use_computer', y='Maths_Percentage', data=df1)
plt.title("Maths Scores vs Computer Usage")
plt.show()
```

```python
sns.heatmap(df1[['Maths_Percentage', 'Reading_Percentage', 'Science_Percentage', 'Social_Percentage']].corr(), annot=True, cmap='Blues')
```

```python
# Maths
sns.barplot(x='gender', y='Maths_Percentage', data=df1, color='red')
plt.title("Average Maths Percentage by Gender")
plt.show()

# Social
sns.barplot(x='gender', y='Social_Percentage', data=df1, color='skyblue')
plt.title("Average Social Percentage by Gender")
plt.show()

# Science
sns.barplot(x='gender', y='Science_Percentage', data=df1, color='green')
plt.title("Average Science Percentage by Gender")
plt.show()
```
